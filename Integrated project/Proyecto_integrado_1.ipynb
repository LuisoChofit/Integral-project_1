## Conclusiones y siguientes pasos

**Conclusiones**

Como se puede observar, los nombres de las columnas empiezan con mayúsculas y la segunda palabra no siempre empieza con mayúscula, hay muchos datos ausentes en los campos de puntuación de los juegos y en la columna `Rating`, y podemos ver que los valores de la columna `User_Score` son de tipo `object` cuando deberían ser de tipo coma flotante.

En el siguiente paso del análisis cambiaremos los nombres de las columnas, cambiándo los nombres todo a minúsculas para facilitar el análisis, convertiremos los datos con tipo incorrecto al tipo de datos adecuado y se estudiará la forma de tratar los valores ausentes.

# Preparar los datos

Para una mayor facilidad del manejo de los datos aplicaremos un formato más uniforme al DataFrame, como modificar los nombres de las columnas dejándolos todo en minúsculas y aplicando los tipos de datos adecuados a aquellas columnas con tipos inadecuados.

Después, se tratarán los valores ausentes y se insertará una nueva columna con el total de las ventas de cada videjuego.

## Reemplazar los nombres de las columnas

Las columnas están escritas con la primera letra en mayúscula y también algunas palabras de aquellas columnas con nombres compuestos de varias palabras. Le aplicaremos a todas un formato con todas las letras minúsculas, lo cual facilitará la escitura del código para procesar el análisis.

```python
# Reemplaza los nombres de columnas con todo minusculas

games_data = games_data.rename(columns={'Name': 'name',
                               'Platform': 'platform',
                               'Year_of_Release': 'year_of_release',
                               'Genre': 'genre',
                               'NA_sales': 'na_sales',
                               'EU_sales': 'eu_sales',
                               'JP_sales': 'jp_sales',
                               'Other_sales': 'other_sales',
                               'Critic_Score': 'critic_score',
                               'User_Score': 'user_score',
                               'Rating': 'rating'})
```

## Convertir los datos a tipos necesarios

En la columna `user_score` encontramos que tiene valores de tipo `objeto`. Convertiremos estos valores en tipo numéricos ya que se puntúa con un valor numérico, al cual se le pueden aplicar métodos estadísticos.

```python
# Cambia los tipos de datos

games_data['user_score'] = pd.to_numeric(games_data['user_score'], errors='coerce')
```

En la columna `user_score` nos encontramos el valor "TBD" (_to be determined_) que significa "a determinar". Con el parámetro `errors='coerce'` cambiamos por "NaN" los valores que no se pueden pasar a tipo numérico.

Se ha utilizado este método ya que si la puntuación está por determinar quiere decir que aún no disponemos de ella y por lo tanto es lo mismo que si fuera un valor ausente.

## Tratar valores ausentes

Encontramos valores ausentes en varias columnas, analizaremos de que manera podemos tratarlos y si se pueden rellenar.

Empezaremos por dos entradas en las columnas `name` y `genre` en las que hay valores ausentes. Comprobaremos primero si coinciden para las mismas filas.

```python
# Comprueba si ambas entradas coinciden

games_data[games_data['name'].isna()]
```

Se puede ver que estas entradas apenas aportan información ya que carecen tanto de nombre como de género, tuvieron muy pocas ventas y además no hay datos sobre puntuación ni clasificación. Estas entradas podemos simplemente eliminarlas.

```python
# Elimina las entradas con valores ausentes en 'name' y 'genre'
games_data = games_data.dropna(subset=['name', 'genre'])
```

Por otro lado tenemos valores ausentes en las columnas `year_of_release`, `critic_score`, `user_score` y `rating`. Estos valores son algo que no se puede rellenar de forma arbitraria mediante algún tipo de correlación, ya que el año de lanzamniento es independiente de cada juego y no se puede obtener de forma exacta a partir de ningnún otro parámetro; y lo mismo sucede para los valores de puntuación, ya que eso es algo subjetivo en base a la experiencia de juego de las compañías que lo puntúan y los usuarios que los juegan.
Por otro lado tampoco se puede rellenar el campo de la clasificación del juego ya que es un valor otorgado por la ESRB que determina, según sus criterios, la clasificación del juego.

**Conclusiones intermedias**

Hay valores ausentes en la columna `rating` que se pueden justificar ya que el organismo ESRB se creó en septiembre de 1994, por lo cual, aquellos juegos anteriores a esa fecha no tendrán clasificación según la ESRB.

Los valores ausentes en las columnas de puntuación se pueden deber a una baja popularidad del videojuego cuando salió o a que sean demasiado antiguos, y por aquella época no se registrasen tanto las valoraciones de los juegos.

## Ventas totales

Tras procesar los datos, añadimos la columna de ventas totales mediante la suma de las 4 columnas de ventas. Llamaremos a la columna `total_sales`.

```python
# Añade la columna 'total_sales'

games_data['total_sales'] = (games_data['na_sales'] + games_data['eu_sales'] +
                             games_data['jp_sales'] + games_data['other_sales'])
```

### Como afectan las reseñas en las ventas

Otro dato que tenemos en nuestro DataFrames son las reseñas tanto de profesionales como de los usuario, aunque no en todos los juegos.

En este apartado analizaremos de todo el conjunto de datos como afecta la puntuación de los juegos de una plataforma de las más populares. Para ser más precisos en este análisis de cara al año 2017, utilizaremos como plataforma de referencia la PlayStation 4, ya que es la más popular y la que más juegos exclusivos tiene.

Veremos como pueden afectar las reseñas mediante un gráfico de dispersión.

```python
# Imprime un gráfico de dispersión de las ventas según las reseñas profesionales

games_data.query('platform == "PS4"').plot(x='critic_score', y='total_sales', kind='scatter', figsize=(12,6))

# obtiene la correlación de la distribución
games_data.query('platform == "PS4"')[['critic_score', 'total_sales']].corr()
```

```python
# Imprime un gráfico de dispersión de las ventas según las reseñas de los usuarios

games_data.query('platform == "PS4"').plot(x='user_score', y='total_sales', kind='scatter', figsize=(12,6))

# obtiene la correlación de la distribución
games_data.query('platform == "PS4"')[['user_score', 'total_sales']].corr()
```

**Conclusiones intermedias**

Las ventas de los juegos se ven más afectadas por las notas que les dan las compañías que prueban los juegos que por aquellas puntuaciones que les dan los usuarios, ya que existe una correlación positiva en el gráfico en el que se comparan las ventas con las reseñas de críticas profesionales, mientras que en el gráfico en el que se comparan las ventas totales con las reseñas de los usuarios no existe una correlación aparente.

Es posible que se deba a que las primeras reseñas que se conocen suelen ser de compañías profesionales antes que las de los usuarios.

### Variación de ventas para un mismo juego en distintas plataformas

Es habitual que un mismo juego se publique en varias plataformas, en este apartado del análisis estudiaremos como varían las ventas de algunos de los juegos más populares en distintas plataformas en el período en el que estamos trabajando.

Primero buscaremos los juegos más populares de la plataforma más popular y comprobaremos si son juegos que han sido publicados en otras plataformas. De los juegos que obtengamos, crearemos un DataFrame que contenga los 10 juegos más populares con los datos de ventas en sus distintas plataformas, a partir del cual obtendremos un gráfico mostrando las ventas por plataforma.

```python
# Filtra los 10 juegos más populares en PlayStation 4

games_data_new = games_data.query('year_of_release > 2014').sort_values(by='name')                    # Filtra y ordena los juegos de 2015 y 2016
games_data_new['multiplatform'] = games_data_new.duplicated('name')                                   # Añade la etiqueta True si hay más entradas de un mismo juego, pero en otra plataforma
top_sales = (games_data_new.query('platform == "PS4" and total_sales > 2 and multiplatform == True')  # Filtra por los juegos de PS4 más populares
             .sort_values(by='total_sales', ascending=False).reset_index(drop=True).head(10))
```

```python
# Amplía la lista con el resto de entradas para un mismo juego en diferentes plataformas

lista = top_sales['name'].unique()

multiplat_games = top_sales

for i in range(len(lista)):    # Añade las demás entradas con los datos de los juegos en otras plataformas
    multiplat_games = multiplat_games.append(games_data_new[games_data_new['name'].str.contains(lista[i])])

multiplat_games = multiplat_games.append(games_data_new[games_data_new['name'].str.contains('Battlefront')]) # No añadía estos juegos
multiplat_games = multiplat_games.drop_duplicates().sort_values('name').set_index(['name','platform'])
```

```python
# Imprime un gráfico con la variación de ventas

(multiplat_games.pivot_table(index='name', columns='platform', values='total_sales')
 .plot(kind='bar', figsize=(20,10), xlabel='Juego', ylabel='Ventas (en millones de USD)'))
plt.title('Variación de ventas de los juegos más populares en distintas plataformas')
```

En el caso de los juegos multiplataforma más populares encontramos que obtuvieron más ventas en PlayStation 4 que en Xbox One, PC o en la generación pasada de consolas de Sony y Microsoft.

**Conclusiones intermedias**

La plataforma en la que más juegos se venden es la PlayStation 4 con diferencia del resto. Puede deberse a que hay más usuarios que tengan una PS4 a que tengan una Xbox One, lo que la hace la plataforma más rentable tanto para juegos multiplataforma como para juegos exclusivos.

# Prueba de hipótesis

Finalmente probaremos la hipótesis formuladas al inicio. Estas hipótesis son las de que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas y la de que las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

## Primera hipótesis. Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.

En este caso supondremos como hipótesis nula la afirmación de que las calificaciones promedio de los usuarios para las plataformas Xbox One y PC no difiere: 

\begin{equation}
H_0: µ_1\neqµ_2
\end{equation}

Por lo tanto, la hipótesis alternativa será que éste sí difiere:

\begin{equation}
H_1: µ_1=µ_2
\end{equation}


Comparamos las calificaciones promedio de los juegos en ambas plataformas en los últimos 2 años, dónde `xbox_score` es el conjunto de calificaciones de los usuarios en los últimos 2 años en la plataforma Xbox One y `pc_score` es el conjunto de califiaciones de los usuarios en PC.

Establecemos el nivel de significancia (`alpha`) en 0.05%. 

```python
xbox_score = games_data_new.query('platform == "XOne"')['user_score'].dropna() #Los valores nulos no interesan
pc_score = games_data_new.query('platform == "PC"')['user_score'].dropna()

alpha = 0.05

results = st.ttest_ind(xbox_score, pc_score)

print('p-value: ', results.pvalue)
if results.pvalue < alpha:
    print('Rechazamos la hipótesis nula')
else:
    print("No rechazamos la hipótesis nula")
```

La probabilidad de que las calificaciones promedio de los usuarios sean diferentes es de un 0.27%, que está bastante por encima del umbral `alpha`, por tanto, no podemos rechazar la hipótesis nula. En este caso diremos que **no aceptamos la hipótesis sobre que las calificaciones promedio entre Xbox One y PC sean las mismas**.

## Segunda hipótesis. Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes difieren.

En este caso supondremos como hipótesis nula la afirmación de que las calificaciones promedio de los usuarios para los géneros de Acción y Deportes no difiere: 

\begin{equation}
H_0: µ_1=µ_2
\end{equation}

Por lo tanto, la hipótesis alternativa será que éste sí difiere:

\begin{equation}
H_1: µ_1\neqµ_2
\end{equation}


Comparamos las calificaciones promedio de los juegos de Acción y Deportes, dónde `action_score` es el conjunto de calificaciones de los usuarios en juegos de género Acción y `sports_score` es el conjunto de calificaciones de los usuarios en juegos de Deportes.

Establecemos el nivel de significancia (`alpha`) en 0.05%. 

```python
action_score = games_data_new.query('genre == "Action"')['user_score'].dropna() #Los valores nulos no interesan
sports_score = games_data_new.query('genre == "Sports"')['user_score'].dropna()

alpha = 0.05

results = st.ttest_ind(action_score, sports_score)

print('p-value: ', results.pvalue)
if results.pvalue < alpha:
    print('Rechazamos la hipótesis nula')
else:
    print("No rechazamos la hipótesis nula")
```

La probabilidad de que las calificaciones promedio de los usuarios entre los géneros de Acción y Deportes sea la misma es ínfima, de este modo aceptamos la hipótesis que planteamos al principio de que **la calificación promedio de los géneros de Acción y Deportes es distinta**.
